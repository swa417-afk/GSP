# Glass Substrate Demo Notebook

Quick start for interacting with the demo modules.

```python
from gsp_core import Engine, Storage, Verification
from gsp_core.reconstruction import Reconstruction

storage = Storage()
engine = Engine(storage=storage, verifier=Verification())
reconstruction = Reconstruction()
```

```python
storage.clear()
engine.submit(
    model_id="demo-model",
    input_hash="input123",
    output_hash="output456",
    policy_id="policy-001",
    actor="notebook",
)
engine.submit(
    model_id="demo-model",
    input_hash="input789",
    output_hash="outputxyz",
    policy_id="policy-002",
    actor="notebook",
)

records = engine.list_records()
status = engine.integrity_status()
latest = reconstruction.latest(records)

print(f"records: {len(records)} | ok: {status['ok']}")
print("latest:", latest)
```
